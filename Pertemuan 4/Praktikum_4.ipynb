```python
import pandas as pd

data = pd.DataFrame({
 'IPK': [3.5, 2.1, 3.8, 2.5, 3.2, 2.0, 3.6, 2.7],
 'Kehadiran': [90, 60, 95, 70, 85, 50, 92, 75],
 'Jam_Belajar': [15, 5, 20, 8, 12, 4, 18, 9],
 'Jumlah_Organisasi': [1, 0, 2, 1, 1, 0, 2, 1],
 'Lulus': [1, 0, 1, 0, 1, 0, 1, 0]
})
print(data)
```

```python
X = data[['IPK', 'Kehadiran', 'Jam_Belajar', 'Jumlah_Organisasi']]
y = data['Lulus']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
 X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
model.fit(X_train_scaled, y_train)
print("Model berhasil dilatih!")
```

```python
prediksi = model.predict(X_test_scaled)
print("Hasil prediksi:", prediksi)
```

```python
from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test, prediksi)
print("Akurasi model:", akurasi)
```
