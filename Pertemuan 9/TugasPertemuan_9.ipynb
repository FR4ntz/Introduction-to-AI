### 1. Baca dataset (20 data)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# 1. Baca dataset (20 data)
data = pd.read_csv("dataset_lengkap.csv").head(20)
```

### 2. Fitur: Jenis Rekomendasi | Label: Impulse Buying

```python
X = data['2. Jenis rekomendasi mana yang paling sering membuat Anda tertarik untuk mengklik produk tersebut? (Pilih satu yang paling dominan)  ']
y = data['4. Apakah Anda pernah membeli produk HANYA karena produk tersebut muncul di kolom rekomendasi (padahal awalnya tidak berniat membeli)?  ']

```

### 3. Preprocessing: Lowercase

```python
X = X.astype(str).str.lower()

```

### 4. Ubah teks ke angka (Vectorization)

```python
vectorizer = CountVectorizer()
X_vector = vectorizer.fit_transform(X)
```

### 5. Split data (80% train, 20% test)

```python
X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size=0.2, random_state=42)
```

### 6. Model Naive Bayes

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

### 7. Prediksi

```python
hasil = model.predict(X_test)
print("Hasil prediksi:", hasil)
```
